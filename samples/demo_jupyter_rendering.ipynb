# python-sec Jupyter Rendering Demo

This notebook demonstrates the `_repr_html_()` methods on all response models.
In Jupyter, placing an object as the last expression in a cell will automatically
render its HTML table representation — no `print()` or `display()` needed.

**Models covered:** `Filing`, `CompanyInfo`, `Submission`, `Fact`, `Facts`, `SearchResult`

## 1. Setup — Import and Create Client

```python
from edgar.client import EdgarClient

# SEC EDGAR requires a User-Agent header.
edgar_client = EdgarClient(user_agent="Your Name your-email@example.com")
```

## 2. Filing — Single Filing Card

Place a `Filing` object as the last expression. Jupyter calls `_repr_html_()` and
renders it as a styled table with form type, date, accession number, and a clickable URL.

```python
# Get Apple's 10-K filings and display the first one.
company = edgar_client.company("AAPL")
filings = company.get_filings(form="10-K")
filings[0]  # Auto-renders as HTML table in Jupyter
```

## 3. CompanyInfo — Company Metadata Card

`CompanyInfo` renders company details (CIK, tickers, SIC, exchange) plus a
sub-table of recent filings.

```python
info = company.get_info()
info  # Renders company card + recent filings table
```

## 4. Submission — Individual Filing Record

Each `Submission` from the `recent_submissions` list renders as its own card.

```python
# Display the most recent submission from Apple's filing history.
info.recent_submissions[0]
```

## 5. Facts — XBRL Taxonomy Overview

`Facts` renders entity info plus a taxonomy summary table showing concept counts.

```python
facts = company.get_facts()
facts  # Entity info + taxonomy concept counts
```

## 6. Fact — Single XBRL Data Point

An individual `Fact` renders as a compact card with value, period, form, and filing details.

```python
# Get Revenue facts and display the most recent one.
revenue_facts = facts.get("us-gaap", "Revenues")
revenue_facts[-1]  # Latest revenue fact as HTML card
```

## 7. SearchResult — Full-Text Search

`SearchResult` renders as a card with company, form, dates, and a clickable filing link.

```python
results = edgar_client.search("revenue recognition", form_types=["10-K"], size=3)
results[0]  # First search result as HTML card
```

## 8. Collections as DataFrames

For lists of objects, use `to_dataframe()` to get a pandas DataFrame that also
auto-renders as a table in Jupyter. Requires `pip install python-sec[pandas]`.

```python
from edgar.models import to_dataframe

# Convert filings list to a DataFrame.
to_dataframe(filings)
```

```python
# Revenue facts as a DataFrame — great for time-series analysis.
facts.to_dataframe("us-gaap", "Revenues")
```

## 9. CSS Styling

All `_repr_html_()` output uses inline CSS so tables render consistently across
Jupyter Lab, Jupyter Notebook, VS Code, and Google Colab.

The styles are defined as module-level constants in `edgar.models`:
- `_TABLE_STYLE` — border-collapse, font family, max-width
- `_TH_STYLE` — header background, padding
- `_TD_STYLE` — cell padding, border
- `_CAPTION_STYLE` — model name label above the table

These are intentionally simple and inline so they work without custom CSS files.
